```python
unique_athletics = athlete_Event_Results.loc[(athlete_Event_Results.sport == "Athletics"), "event"].unique()

women_athletics = []
items_to_remove = []

i = 0

for s in unique_athletics:
    if "Women" in s:
        women_athletics.append(s)
        items_to_remove.append(i)
    i += 1

unique_athletics = np.delete(unique_athletics, items_to_remove)

display(women_athletics)
```
